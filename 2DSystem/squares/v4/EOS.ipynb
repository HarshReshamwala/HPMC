```python
import json
import numpy as np
import matplotlib.pyplot as plt

# Load data from JSON files
with open('simulation_40_to_20.json', 'r') as f:
    data_40_to_20 = json.load(f)

with open('simulation_0_to_20.json', 'r') as f:
    data_20_to_0 = json.load(f)

# Extract arrays
pressures_40_to_20 = np.array(data_40_to_20['pressures'])
avg_vf_40_to_20 = np.array(data_40_to_20['avg_vf'])
std_vf_40_to_20 = np.array(data_40_to_20['std_vf'])

pressures_20_to_0 = np.array(data_20_to_0['pressures'])
avg_vf_20_to_0 = np.array(data_20_to_0['avg_vf'])
std_vf_20_to_0 = np.array(data_20_to_0['std_vf'])

# Plot the first dataset (from 40 to 20) in blue
plt.plot(avg_vf_40_to_20, pressures_40_to_20, linestyle='-', marker='o', color='b', label="Avg VF (40 to 20)")
plt.errorbar(avg_vf_40_to_20, pressures_40_to_20, xerr=std_vf_40_to_20, fmt='o', capsize=5, color='b')

# Plot the second dataset (from 20 to 0) in red, excluding the 20 pressure to avoid connecting the lines
plt.plot(avg_vf_20_to_0[1:], pressures_20_to_0[1:], linestyle='-', marker='o', color='r', label="Avg VF (20 to 0)")
plt.errorbar(avg_vf_20_to_0[1:], pressures_20_to_0[1:], xerr=std_vf_20_to_0[1:], fmt='o', capsize=5, color='r')

# Add labels, title, and legend
plt.ylabel("Pressure (beta_P)")
plt.xlabel("Average Volume Fraction")
plt.title("Average Volume Fraction vs. Pressure with Error Bars")
plt.grid()
plt.legend()

# Show plot
plt.show()
```
